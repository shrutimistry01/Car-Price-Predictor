```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv("cleaned_data.csv")
final_dataset= pd.read_csv("cleaned_numerical_data.csv")
df.head()
```

```python
final_dataset.head()
```

```python
corr_matrix=final_dataset.corr()
corr_matrix
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
```

## categorical data

```python
df_categorical=[ 'fuel','seller_type', 'transmission', 'owner', 'seats', 'company_name', 'no_of_years']
for i in df[df_categorical].columns:
    cat_vc=df[i].value_counts()
    print("Graph for {} : total categories= {}".format(i, len(cat_vc)))
    sns.barplot(x =cat_vc.index ,y = cat_vc)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# independent and dependent feature respectively
X=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
```

## Feature Importance

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
feat_imps=pd.Series(model.feature_importances_,index=X.columns)
feat_imps.nlargest(10).plot(kind="barh")
plt.show()
```
