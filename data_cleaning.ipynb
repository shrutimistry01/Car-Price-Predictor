```python
import pandas as pd
```

```python
df=pd.read_csv("Car details.csv")
df.head()
```

```python
df.columns
```

### Categorical data

```python
cat_col=['fuel', 'seller_type','transmission', 'owner']
for i in cat_col:
    print(i,df[i].unique())
```

## Conveting non-numeric to numeric data

```python
#removing kmpl from mileage

mil_fix= df['mileage'].apply(lambda x : str(x))
mil_fix= mil_fix.apply(lambda x : x.replace(" ","").replace("kmpl","").replace("km/kg",""))
df["mileage"]= mil_fix.apply(lambda x : float(x))
```

```python
df.head()
```

```python
#removing CC from engine

eng_fix= df['engine'].apply(lambda x : str(x))
eng_fix= eng_fix.apply(lambda x : x.replace(" ","").replace("CC",""))
df["engine"]= eng_fix.apply(lambda x : float(x))
df.head()
```

```python
#removing bhp from max_power

import numpy as np

mp_fix = df["max_power"].apply(lambda x : str(x))
mp_fix = mp_fix.apply(lambda x : x.replace("bhp",""))
mp_fix = pd.to_numeric(mp_fix, errors='coerce')
df["max_power"] = mp_fix.replace(np.nan, 0, regex=True)
#df["max_power"]= mp_fix.apply(lambda x : float(x))
#pd.to_numeric(df["max_power"])
df.head()
```

```python
df.drop('torque',axis=1,inplace=True)
```

```python
df.head()
```

```python
# check missing values
df.isnull().sum()
```

```python
#handling missing value
df = df.fillna(0)
```

```python
df.info()
```

###  Adding column 'company_name' 

```python
df['company_name']=df["name"].apply(lambda x : x.split(" ",1)[0])
df.head()
```

```python
df.rename(columns={'name': 'model_name'}, inplace=True)
```

```python
currentYear=2022
df["no_of_years"]= currentYear-df.year
```

```python
df.head()
```

```python
df.describe().apply(lambda s: s.apply('{0:.5f}'.format))# df.describe()
```

```python
df.head()
```

### create dummies

```python
df1= pd.get_dummies(df[cat_col],drop_first=True)
final_dataset=pd.concat([df,df1],axis=1)
```

### drop non numeric data in final dataset

```python
final_dataset.drop(final_dataset[cat_col],inplace=True,axis=1)
final_dataset.drop(final_dataset[['year','model_name','company_name']],inplace=True,axis=1)
final_dataset.head()
```

```python
final_dataset.to_csv("cleaned_numerical_data.csv",index=False)
```
