```python
import pandas as pd
```

```python
df=pd.read_csv("Car details.csv")
df.head()
```

```python
df.columns
```

### Categorical data

```python
cat_col=['fuel', 'seller_type','transmission', 'owner']
for i in cat_col:
    print(i,df[i].unique())
```

## Conveting non-numeric to numeric data

```python
#removing kmpl from mileage

mil_fix= df['mileage'].apply(lambda x : str(x))
mil_fix= mil_fix.apply(lambda x : x.replace(" ","").replace("kmpl","").replace("km/kg",""))
df["mileage"]= pd.to_numeric(mil_fix,errors='coerce')
```

```python
df.head()
```

```python
#removing CC from engine

eng_fix= df['engine'].apply(lambda x : str(x))
eng_fix= eng_fix.apply(lambda x : x.replace(" ","").replace("CC",""))
df["engine"]= pd.to_numeric(eng_fix,errors='coerce')
df.head()
```

```python
#removing bhp from max_power

import numpy as np

mp_fix = df["max_power"].apply(lambda x : str(x))
mp_fix = mp_fix.apply(lambda x : x.replace("bhp",""))
mp_fix = pd.to_numeric(mp_fix, errors='coerce')
df["max_power"] = mp_fix.replace(np.nan, 0, regex=True)
#df["max_power"]= mp_fix.apply(lambda x : float(x))
#pd.to_numeric(df["max_power"])
df.head()
```

```python
df.drop('torque',axis=1,inplace=True)
```

```python
df.head()
```

```python
# check missing values
df.isnull().sum()
```

```python
#handling missing value
# from sklearn.impute import SimpleImputer
# impute= SimpleImputer(strategy="median")

#for mileage
impute_m= df.mileage.median()
df["mileage"] = df.mileage.fillna(round(impute_m,2))
print(impute_m)
```

```python
#for engine
impute_e= df.engine.median()
df["engine"] = df.engine.fillna(round(impute_e,0))
print(impute_e)
```

```python
df.seats.value_counts()
```

```python
#for Seat by taking mode which is 5
df["seats"] = df.seats.fillna(5.0)

df.seats.isnull().sum()
```

```python
df.isnull().sum()
```

###  Adding column 'company_name' 

```python
df['company_name']=df["name"].apply(lambda x : x.split(" ",1)[0])
df.head()
```

```python
currentYear=2022
df["no_of_years"]= currentYear-df.year
```

```python
df.drop({'name','year'}, inplace=True,axis=1)
df.head()
```

```python
import seaborn as sns 
from matplotlib import rcParams

rcParams['figure.figsize'] = 15,8
sns.histplot(x="selling_price",hue="owner", data=df,log_scale=True,palette="tab10")
```

```python
mapper={'First Owner':0, 'Second Owner':1, 'Third Owner':2,'Fourth & Above Owner':2, 'Test Drive Car':2}
df.owner =df.owner.replace(mapper)
```

```python
df.describe().apply(lambda s: s.apply('{0:.5f}'.format))# df.describe()
```

```python
df.head()
```

### create dummies

```python
df1= pd.get_dummies(df[cat_col],drop_first=True)
final_dataset=pd.concat([df,df1],axis=1)
```

### drop non numeric data in final dataset

```python
final_dataset.drop(final_dataset[cat_col],inplace=True,axis=1)
final_dataset.drop(final_dataset[['company_name']],inplace=True,axis=1)
final_dataset.head()
```

```python
final_dataset.to_csv("cleaned_numerical_data.csv",index=False)
```
